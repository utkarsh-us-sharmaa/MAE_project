```python
import pandas as pd
```

```python
complete_dataset = pd.read_csv('complete_dataset.csv')
```

```python
complete_dataset.head()
```

```python
results_dataset = pd.read_csv('test_results_by_thickness.csv')
```

```python
results_dataset.head()
```

```python
results_dataset = results_dataset.sort_values('Row_ID')
```

```python
results_dataset
```

```python
results_dataset.to_csv('final_test_result.csv', index=False)
```

```python
decimals = 6  # Adjust this number based on your data precision needs

# Round columns in both dataframes
merge_columns = ['x (Thickness)', 'ModelRatio', 'Tdelay', 'y (Thermal Conductivity Profile)']

results_dataset_rounded = results_dataset.copy()
complete_dataset_rounded = complete_dataset.copy()

for col in merge_columns:
    results_dataset_rounded[col] = results_dataset_rounded[col].round(decimals)
    complete_dataset_rounded[col] = complete_dataset_rounded[col].round(decimals)

# Perform the merge with rounded values
merged_results = pd.merge(
    results_dataset_rounded,
    complete_dataset_rounded[['x (Thickness)', 'ModelRatio', 'Tdelay', 'y (Thermal Conductivity Profile)','Equations', 'Slope', 'Intercept', 'Equation ID']],
    on=['x (Thickness)', 'ModelRatio', 'Tdelay', 'y (Thermal Conductivity Profile)'],
    how='left'
)
```

```python
print("Original results_dataset rows:", len(results_dataset))
print("Merged results rows:", len(merged_results))
```

```python
print("Unique combinations in complete_dataset:")
complete_check = complete_dataset.groupby(['x (Thickness)', 'ModelRatio', 'Tdelay', 'y (Thermal Conductivity Profile)']).size().reset_index(name='count')
print(complete_check[complete_check['count'] > 1])

# Check uniqueness in results_dataset
print("\nUnique combinations in results_dataset:")
results_check = results_dataset.groupby(['x (Thickness)', 'ModelRatio', 'Tdelay', 'y (Thermal Conductivity Profile)']).size().reset_index(name='count')
print(results_check[results_check['count'] > 1])
```
